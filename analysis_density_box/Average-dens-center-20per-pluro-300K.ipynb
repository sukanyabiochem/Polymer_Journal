```python
## Libraries require to run the script##import os
import os
import glob
import matplotlib.pyplot as plt
import numpy as np
import sympy as sp
import pandas as pd
from scipy import interpolate
from matplotlib.ticker import ScalarFormatter
import math
from pathlib import Path
from pylab import rc
```

```python
## read data files using pandas ##
dens_EMI = pd.read_csv("density-EMI.xvg",names=["r", "dens"], sep=r'\s+', skiprows=24)
dens_NC = pd.read_csv("density-NC.xvg",names=["r", "dens"], sep=r'\s+', skiprows=24)
dens_EO = pd.read_csv("test-EO.xvg",names=["r", "dens"], sep=r'\s+', skiprows=24)
dens_PO = pd.read_csv("test-PO.xvg",names=["r", "dens"], sep=r'\s+', skiprows=24)
```

```python
# Setup
rc('axes', linewidth=3.0)
fig = plt.figure(1034, figsize=(9, 7))
ax = fig.add_subplot(111)

# Base paths
#base_path_330K = Path("/home/sukanya/Desktop/Solvation-free-energy/phase-seperation-pluronic-emim-dca/15nm-box-60per-IL-40per-pluronic/folder-330K/density-analysis/")
##base_path_300K = Path("/home/sukanya/Desktop/Solvation-free-energy/phase-seperation-pluronic-emim-dca/15nm-box-60per-IL-40per-pluronic/SASA-300K")

plt.plot(dens_EMI["r"]*10, dens_EMI["dens"], color='cyan', linewidth=3.5, label="[EMIM]$^+$")
plt.plot(dens_NC["r"]*10, dens_NC["dens"], color='purple', linewidth=3.5, label="[DCA]$^-$")
plt.plot(dens_EO["r"]*10, dens_EO["dens"], color='gray', linewidth=4, label="EO bead")
plt.plot(dens_PO["r"]*10, dens_PO["dens"], color='magenta', linewidth=4, label="PO bead")   

# Axis labels and ticks
plt.ylabel('Density (kg/m$^3$)', fontsize=40, labelpad=7, fontname='DejaVu Sans')
plt.xlabel('Position from center (Å)', fontsize=40, labelpad=7, fontname='DejaVu Sans')
plt.ylim([0.0, 1000.01])
plt.xlim([-70.01, 70.01])
plt.xticks(np.arange(-70.01, 70.01, 20), fontsize=38, fontname='DejaVu Sans')
plt.yticks(np.arange(0, 1000.02, 200), fontsize=38, fontname='DejaVu Sans')

# Legend and formatting
plt.legend(fontsize=25, bbox_to_anchor=(0.79, 1.025), loc='upper right', edgecolor='white')
plt.tick_params(axis='both', which='major', length=18, direction='in', width=2.5, labelsize=35, pad=10)

# Save and show
plt.savefig("Figure7A.png", bbox_inches='tight', dpi=600)
plt.show()
```
